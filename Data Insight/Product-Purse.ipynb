```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset

df = pd.read_csv(r"C:\Users\win10\Documents\Prohect_Sell SMart.csv")

# Preview data
print(df.head())
```

```python

print(df[["Original Price", "Discounted Price", "Rating"]].describe())
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=df["Brand"], order=df["Brand"].value_counts().index, palette="coolwarm")
plt.title("Brand-wise Sales Distribution", fontsize=14)
plt.xlabel("Number of Sales")
plt.ylabel("Brand")
plt.show()

```

```python
plt.figure(figsize=(25, 10))
sns.scatterplot(x=df["Original Price"], y=df["Discounted Price"], hue=df["Brand"], alpha=0.7, palette="Set1")
plt.title("Price vs Discount Cost")
plt.xlabel("Original Price")
plt.ylabel("Discounted Price")
plt.show()

```

```python
print(df.dtypes)
```

```python
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
print(df["Rating"].isna().sum())  # Count missing values
print(df["Rating"].value_counts())  # Check unique values
```

```python
plt.figure(figsize=(8, 5))
df["Rating"].plot(kind="hist", bins=20, color="blue", edgecolor="black")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```
